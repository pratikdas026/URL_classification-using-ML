```python
import numpy as np
import pandas as pd
import whois
```

```python
urldata = pd.read_csv('/content/drive/MyDrive/benign-phishing-url-classification-using-whois-and-lexical-features-master/Data/CleanedDataset/preprocessed_urldata(short).csv',encoding = "ISO-8859-1")
```

```python
len(urldata)
```

```python
#this function performs whois and returns true to those urls which are reachable through whois
def performwhois(url):
    try:
        result = whois.whois(url)
        return True #success
    except Exception:
        return False #error
```

```python
benign_sample = urldata['domain'][252:503]
```

```python
len(benign_sample)
```

```python
type(benign_sample)
```

```python
benign_sample.head()
```

```python
benign_urls = []
counter = 0
for url in benign_sample:
    if performwhois(url):
        counter = counter + 1
        print(counter)
        benign_urls.append(url)
```

```python
len(benign_urls)
```

```python
benign_df = pd.DataFrame()
```

```python
benign_df['url'] = benign_urls
```

```python
benign_df['label'] = 0
```
